# Import Series with the read_csv Method

```python
import pandas as pd
```

```python
# By default function read_csv create a DataFrame 
pokemon = pd.read_csv('../../CompletedCourseFiles/Pandas CSV Files/pokemon.csv', usecols = ['Pokemon'])
pokemon
```

```python
# With parameter squeeze = True we created a Series and don't more a DataFrame
google = pd.read_csv('../../CompletedCourseFiles/Pandas CSV Files/google_stock_price.csv', squeeze = True)
google
```
